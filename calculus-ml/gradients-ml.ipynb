```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import math
```

Define the function z = x² - y²

```python
def function(my_x, my_y):
    return my_x**2 - my_y**2
```

plotting z with respect to x by varying x...

```python
xs = np.linspace(-3, 3, 1000)
```

while holding y as a constant(e.g. y = 0)

```python
zs_wrt_x = function(xs, 0)
```

```python
fig, ax = plt.subplots()

plt.axvline(x=0, color="lightgray")
plt.axhline(y=0, color="lightgray")

plt.xlabel("x")
plt.ylabel("z")

_ = ax.plot(xs, zs_wrt_x)
```

To find the slope of z w.r.t. x in a given point, I'll use the partial derivate from $z = x² - y²$, that will be: 
$\partial z/ \partial x = 2x$

```python
def delz_delx (my_x, my_y): # y won't be used in this particular case, but it can be on more elaborate cases tho
    return 2 * my_x
```

```python
def point_tangent_wrt_x(my_xs, my_x, my_y, my_func, func_prime, color):
    my_z = my_func(my_x, my_y)
    plt.scatter(my_x, my_z, c=color, zorder=3)
    
    tangent_m = func_prime(my_x, my_y)
    tangent_b = my_z - tangent_m * my_x
    tangent_line = tangent_m * my_xs + tangent_b
    
    plt.plot(my_xs, tangent_line, c=color,
             linestyle="dashed", linewidth=0.7,zorder=3)
```
