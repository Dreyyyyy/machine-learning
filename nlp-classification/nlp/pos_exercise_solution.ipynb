```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

<h3>Read a new story</h3>

```python
with open("./news_story.txt", "r") as file:
    data = file.read()
    
doc = nlp(data)
doc
```

<h3>Extract NOUN and NUM tokens</h3>

```python
noun_list = []
num_list = []

for token in doc:
    if token.pos_ == "NOUN":
        noun_list.append(token.text)
    elif token.pos_ == "NUM":
        num_list.append(token.text)
```

```python
print(num_list)
```

```python
print(noun_list)
```

<h3>Print a count of all POS tags</h3>

```python
for token in doc:
    print(f"{token.pos_} : {token.pos}")
```
