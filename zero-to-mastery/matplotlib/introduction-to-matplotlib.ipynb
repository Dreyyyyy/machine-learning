## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [10, 20, 30, 40]
y = [100, 200, 300, 400]
     
# 2. Setup plot
fig, ax = plt.subplots(figsize=(8, 8))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot;
* Scatter plot;
* Bar plot;
* Histogram;
* Subplots.

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:100]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(),
       nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price($)")

plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
plt.show()
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, 
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

plt.show()
```

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("../pandas/data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2024", periods=1000))
ts = ts.cumsum()
ts.plot()
plt.show()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace(r"[$,.]", "", regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2].astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales");
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(car_sales["Sale Date"],
        car_sales["Total Sales"])

ax.set_xlabel("Date")
ax.set_ylabel("Total Sales")
ax.set_title("Sales Date vs. Total Sales")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(car_sales["Odometer (KM)"],
           car_sales["Price"])

ax.set_xlabel("Odometer (KM)")
ax.set_ylabel("Price")
ax.set_title("Odometer vs. Price")

plt.show()
```

```python
heart_disease = pd.read_csv("./data/heart-disease.csv")
heart_disease
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(heart_disease["age"])
ax.set_title("Age Histogram")
ax.set_xlabel("Age")
ax.set_ylabel("Number of Patients")

plt.show()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 30));
plt.show()
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target")

plt.show()
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10, 6))

over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)

ax.set_xlim([45, 100])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
           linestyle="--")

plt.show()
```

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```
