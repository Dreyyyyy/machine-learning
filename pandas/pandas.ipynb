```python
import pandas as pd
```

Read a CSV file into a Pandas DataFrame and display the first few rows.

```python
df = pd.read_csv("/home/shuruyi/Documents/git/machine-learning/pandas/sample_file.csv")
df.head()
```

Create a new column in the DataFrame by combining values from existing columns.

```python
df = df.assign(FULL_ADDRESS=lambda x: x['STREET_NAME'] + ", " + x["STATE"])
df
```

Filter the DataFrame to select rows based on a condition (e.g., values greater than a certain threshold).

```python
df_filtered = df[df["LATITUDE"] > 5]
df_filtered
```

Sort the DataFrame based on a specific column in ascending order.

```python
df_ascending = df.sort_values(by=["LATITUDE"], ascending=True)
df_ascending
```

Calculate summary statistics (mean, median, min, max) for numerical columns in the DataFrame.

```python
df.groupby("STATE")[["LATITUDE"]].agg(["mean", "median", "min", "max"])
```

Group the DataFrame by a categorical column and calculate aggregate functions (e.g., sum, mean) for each group.

```python
df.groupby("COUNTY")[["LATITUDE"]].agg(["mean", "sum", "median", "min", "max"])
```

Merge two DataFrames based on a common column.

```python
df_merged = df.merge(df_ascending, on="ZIP_CODE")
df_merged
```

Remove duplicate rows from the DataFrame.

```python
df.drop_duplicates()
df
```

Handle missing data by filling NaN values with a specific value or using interpolation.

```python
mean_of_sums = df.select_dtypes(include=['number']).sum().mean()
df = df.fillna(mean_of_sums)
df
```

Export the DataFrame to a CSV file.

```python
df.to_csv("/home/shuruyi/Documents/git/machine-learning/pandas/sample_file_export.csv")
```
