## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) #width, height

# 3. Plot data
ax.plot(x, y)

# 4. Cutomize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("/home/shuruyi/Desktop/git/machine-learning/sample_project/images/sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
plt.show()
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

#  Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

```python
import pandas as pd

heart_disease = pd.read_csv("/home/shuruyi/Desktop/git/machine-learning/sample_project/heart-disease.csv")
heart_disease.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

heart_disease["age"].plot.hist()
plt.show()
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True)
plt.show()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target")
plt.show()
```

```python
# OO Method mixed with Pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
plt.show()
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax.set(title="Heart disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
           linestyle='--')

plt.show()
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

ax0.set(title="Heart disease and Cholesterol levels",
        ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

ax1.set(title="Heart disease and Thalach levels",
        xlabel="Age",
        ylabel="Talach")

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```
