```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create a line plot showing the variation of temperature over a week.

```python
temp_data = np.random.randint(20, 39, size=7) 
temp_data

days_of_week = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt.plot(temp_data)
plt.title("Temperature variation over a week")
plt.xlabel("Days of the week")
plt.ylabel("Temperature in Celcius")
plt.xticks(range(len(days_of_week)), days_of_week)
plt.yticks(range(18, 41, 2))

plt.show()
```

Plot a scatter plot for random x and y data points.

```python
x_random = np.random.random(40)
y_random = np.random.random(40)

plt.scatter(x_random, y_random)
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Generate a histogram for a given set of data.

```python
plt.hist(x_random)
plt.show()
```

Create a bar chart to compare sales data for different products.

```python
sales_data_num = np.random.uniform(10, 500, size=700)
sales_data_num_rounded = sales_data_num.round(2)

categories = ["Techno", "Utils", "Comestics", "Clothes", "Shoes", "Rings", "Videogames"]
probabilities = [0.45, 0.10, 0.25, 0.05, 0.05, 0.05, 0.05]
sales_data_cat = np.random.choice(categories, size=700, p=probabilities)

sales_data =  pd.DataFrame({"Category": sales_data_cat, "Sales": sales_data_num_rounded})
sales_data_sum = sales_data.groupby("Category")["Sales"].sum()

plt.bar(sales_data_sum.index, sales_data_sum.values)
plt.yticks(range(0, 85000, 7500))
plt.show()
```

Plot a pie chart to represent the distribution of expenses in a budget.

```python
budget_num = np.random.uniform(10, 500, size=500)
budget_features = ["Food", "Entertainment", "Clothing", "Utilities", "Housing", "Transportation", "Insurance", "Healthcare", "Education", "Gifts"]
prob = [0.15, 0.15, 0.20, 0.10, 0.10, 0.10, 0.05, 0.05, 0.05, 0.05]
budget_features_random = np.random.choice(budget_features, size=500, p=prob)

budget_df = pd.DataFrame({"Expenses": budget_features_random, "Budget": budget_num})
budget_df_sum = budget_df.groupby("Expenses")["Budget"].sum()

plt.pie(budget_df_sum.values, labels=budget_df_sum.index)
plt.show()
```

Generate a box plot for a dataset to visualize the distribution of values.

```python
x_random = np.random.normal(0, 1, 1000)
plt.boxplot(x_random)
plt.show()
```

Generate a heatmap to visualize correlation between variables in a dataset.

```python
data = np.random.rand(10,10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()
```

Plot multiple graphs in one figure

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].imshow(data, cmap='hot', interpolation='nearest')
ax[0, 1].hist(x_random)
ax[1, 0].scatter(np.random.randint(0, 20, 100), np.random.randint(20, 60, 100))
ax[1, 1].bar(sales_data_sum.index, sales_data_sum.values)

ax[0, 0].set_title("Heatmap")
ax[0, 1].set_title("Histogram")
ax[1, 0].set_title("Scatter")
ax[1, 1].set_title("Bar")

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.35, wspace=0.35)

plt.show()
```
